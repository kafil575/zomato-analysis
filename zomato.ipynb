```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe.head())
```

```python
def handleRate(value):
	value=str(value).split('/')
	value=value[0];
	return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

```python
dataframe.info()

```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(dataframe['rate'], kde=True, bins=20, color='blue')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=dataframe, x='online_order', y='rate', hue='online_order',palette='Set2')
plt.title('Online Order Availability vs Rating')
plt.xlabel('Online Order')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=dataframe, x='book_table', y='rate', hue='book_table', palette='Set3', dodge=False)
plt.title('Table Booking Availability vs Rating')
plt.xlabel('Table Booking')
plt.ylabel('Rating')
plt.legend([], [], frameon=False)  # Hides the redundant legend
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dataframe, x='votes', y='rate', hue='online_order', alpha=0.7, palette='viridis')
plt.title('Votes vs Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.legend(title='Online Order')
plt.show()
```
